```python
#----- d. Visual Inspection of the Soc_Dem

fig, axs = plt.subplots(3,2, figsize = (15,15))
plt.subplots_adjust(hspace = 0.3, wspace = 0.1)

#----- Feature: Age
axs[0,0].scatter(Soc_Dem['Client'],Soc_Dem['Age'],color = "black",s =10)
axs[0,0].set(xlabel= 'Client_ID', ylabel='Age')

age_count = Soc_Dem['Age'].value_counts()
axs[0,1].bar(age_count.index, age_count, color = 'black' )
axs[0,1].set(xlabel = 'Age', ylabel = 'Frequency')

#----- Feature: Tenure
axs[1,0].scatter(Soc_Dem['Client'],Soc_Dem['Tenure'],color = "black",s =10)
axs[1,0].set(xlabel= 'Client_ID', ylabel='Tenure')

tenure_count = Soc_Dem['Tenure'].value_counts()
axs[1,1].bar(tenure_count.index, tenure_count, color = 'black' )
axs[1,1].set(xlabel = 'Tenure', ylabel = 'Frequency')

#----- Feature: Sex
axs[2,0].scatter(Soc_Dem['Client'],Soc_Dem['Sex'],color = "black",s =10)
axs[2,0].set(xlabel= 'Client_ID', ylabel='Sex')

sex_count = Soc_Dem['Sex'].value_counts()
axs[2,1].bar(sex_count.index, sex_count, color = 'black' )
axs[2,1].set(xlabel = 'Sex', ylabel = 'Frequency')

plt.grid()
plt.show()
```
